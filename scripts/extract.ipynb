```python
from utils import file_to_df, read_prompt, extract_multi_async
import extractors

from functools import partial
import logging, sys

import pandas as pd

logging.basicConfig(
    level=logging.INFO,
    stream=sys.stdout
)
```

```python
loaded_data = file_to_df("../data/input/KerrMcGeeMetadataSetAITest.xlsx")[:5]
```

```python
process = partial(
    extract_multi_async,
    system_prompt=read_prompt("Extract_OCLC"),
    extractor=extractors.OclcFast,
    record_contents=loaded_data["Description (Islandora)"]
)
```

```python
#loaded_data['Subjects_GPT-OSS'] = pd.Series(await process(model="gpt-oss"))
#loaded_data['Subjects_GEMMA4'] = pd.Series(await process(model="gemma4"))
loaded_data['Subjects_QWEN3'] = pd.Series(await process(model="bedrock/qwen.qwen3-coder-30b-a3b-v1:0"))
```

```python
loaded_data.to_csv("../data/output/metadata.csv", index=False)
```
