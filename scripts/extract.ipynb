```python
from utils import file_to_df, read_prompt, extract_multi
from pydantic import BaseModel

class SubjectExtractor(BaseModel):
    subjects: list[str]

loaded_data = file_to_df("../data/input/KerrMcGeeMetadataSetAITest.xlsx")[:5]
```

```python
gpt_oss = extract_multi(
    model="gpt-oss",
    system_prompt=read_prompt("Extract_Subject"),
    extractor=SubjectExtractor,
    record_contents=loaded_data["Description (Islandora)"]
)
```

```python
import pandas as pd
loaded_data['Subjects_GPT-OSS'] = pd.Series(gpt_oss).apply(lambda x: " | ".join(sorted(x.subjects)))
```

```python
loaded_data.to_csv("../data/output/metadata.csv", index=False)
```
